# **Aula 1 - Conociendo la biblioteca Matplotlib**


##**Conociendo los datos**

### Importando los datos

```python
import pandas as pd
```

```python
df = pd.read_csv('inmigrantes_canada.csv')
df.info()
```

### Realizando algunos tratamientos

```python
df.set_index('País', inplace=True)
anos = list(map(str, range(1980, 2014)))
anos
```

### Extrayendo la serie de datos para Brasil

```python
brasil = df.loc['Brasil', anos]
brasil_dict = {'ano': brasil.index.tolist(), 'inmigrantes': brasil.values.tolist()}

datos_brasil = pd.DataFrame(brasil_dict)
datos_brasil
```

```python
import matplotlib.pyplot as plt
```

## Evolución de la inmigración brasileña hacia Canadá

```python
fig, ax = plt.subplots()
ax.plot(datos_brasil['ano'], datos_brasil['inmigrantes'], lw=3)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de Brasil para Canadá entre 1980 a 2013', fontsize=18, loc='left')
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('Número de inmigrantes', fontsize=14)

plt.show()
```
