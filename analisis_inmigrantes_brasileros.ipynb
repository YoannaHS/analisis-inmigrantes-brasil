# **Aula 1 - Conociendo la biblioteca Matplotlib**


##**Conociendo los datos**

### Importando los datos

```python
import pandas as pd
```

```python
df = pd.read_csv('inmigrantes_canada.csv')
df.info()
```

### Realizando algunos tratamientos

```python
df.set_index('País', inplace=True)
anos = list(map(str, range(1980, 2014)))
anos
```

### Extrayendo la serie de datos para Brasil

```python
brasil = df.loc['Brasil', anos]
brasil_dict = {'ano': brasil.index.tolist(), 'inmigrantes': brasil.values.tolist()}

datos_brasil = pd.DataFrame(brasil_dict)
datos_brasil
```

```python
import matplotlib.pyplot as plt
```

## Evolución de la inmigración brasileña hacia Canadá

```python
fig, ax = plt.subplots()
ax.plot(datos_brasil['ano'], datos_brasil['inmigrantes'], lw=3)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de Brasil para Canadá entre 1980 a 2013', fontsize=18, loc='left')
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('Número de inmigrantes', fontsize=14)
ax.set_facecolor('lightgray')

plt.show()
```

```python
# Crear dataframe con países de América del Sur
america_sul = df.query('Região == "América do Sul"').sort_values(by='Total', ascending=True)

# Lista de colores para destacar Brasil
cores = []
for pais in america_sul.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')

# Crear figura
fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul.index, america_sul['Total'], color=cores)

ax.set_title('América del sur: Brasil es el cuarto país con inmigrantes\npara Canadá en el período de 1980 a 2013')
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')

ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

# Agregar valores en las barras
for i, v in enumerate(america_sul['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

# Ajustes visuales
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

# Guardar figura
plt.savefig('images/inmigracion_america_sur.png', dpi=300, bbox_inches='tight')

plt.show()
```
